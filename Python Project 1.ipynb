```python
#import Pandas
import pandas as pd
#read the csv file into a dataframe df
df=pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
#cleaning task - 
#replace 's' with '4' for all columns; as 's' represents values between 0 to 5, taking 4 as one of the values
df.replace('s',4,inplace=True)
```

```python
#convert "Mean Score" column to float data type
df=df.astype({'Mean Score':'float'})
```

```python
#to get the 3rd character from 'School DBN'   
df['School DBN'].str[2:3]
#create new column Borough in df from the 3rd Character 
borough=df['School DBN'].str[2:3]
df['Borough'] = borough
```

```python
#define Boroughs - 
#Brooklyn		K
#Bronx		X
#Queens 		Q
#Manhattan	M
#Staten Island	R
#replace Borough Codes with Borough Names
df['Borough'].replace('K','Brooklyn',inplace=True)
df['Borough'].replace('X','Bronx',inplace=True)
df['Borough'].replace('Q','Queens',inplace=True)
df['Borough'].replace('M','Manhattan',inplace=True)
df['Borough'].replace('R','Staten Island',inplace=True)
```

```python
#convert "Number Scoring Below 65" column to int data type
df=df.astype({'Number Scoring Below 65':'int'})
#convert "Number Scoring Above 65" column to int data type
df=df.astype({'Number Scoring 65 or Above':'int'})
```
